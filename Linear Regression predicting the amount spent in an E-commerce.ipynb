```python
#Lets split the columns:target and the variables
x = df_import[['Avg. Session Length','Time on App','Time on Website','Length of Membership']]
y = df_import['Yearly Amount Spent']
```

```python
#Let's split the data into training and testing
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(x,y, test_size= 0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(X_train,y_train)
```

```python
y_pred = regressor.predict(X_test)
y_pred
```
